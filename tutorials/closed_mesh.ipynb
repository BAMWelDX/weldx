```python
from weldx import get_groove, Q_, CoordinateSystemManager
from weldx.geometry import LinearHorizontalTraceSegment, Trace, Geometry
```

```python
groove = get_groove(
    groove_type="VGroove",
    workpiece_thickness=Q_("0.5 cm"),
    groove_angle=Q_("50 deg"),
    root_face=Q_("1 mm"),
    root_gap=Q_("1 mm"),
)
groove
```

```python
seam_length = 150

trace_segment = LinearHorizontalTraceSegment(seam_length)
trace = Trace(trace_segment)

geometry = Geometry(groove.to_profile(width_default=Q_("50 mm")), trace)
```

```python
groove
```

```python
csm = CoordinateSystemManager("b")
csm.create_cs("1","b", coordinates=[[0, 0, 0], [150, 0, 0]], time=Q_([0, 100], "s"))
csm.assign_data(geometry.spatial_data(5, 5), "data", "b")
```

```python
csm.plot(backend="k3d")
```
