```python
import xarray as xr
import numpy as np
from weldx import SpatialData, CoordinateSystemManager
```

```python
time = ["0s", "3s", "6s", "9s"] 

data = np.array([[[0.,0.,0.],[0.,1., np.sin(i)],[0.,2.,0.]] for i in range(4)])
```

```python
sd = SpatialData(coordinates = data, time = time)
```

```python
csm = CoordinateSystemManager("specimen")
csm.create_cs("scanner", "specimen", coordinates=[[0,0,0], [15,0,0]], time=["0s","15s"])
```

```python
csm.assign_data(sd, "scan_data","scanner")
```

```python
csm.plot()
```
